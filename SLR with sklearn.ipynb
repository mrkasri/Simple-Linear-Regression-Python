## Simple Linear Regression (sklearn) الارتباط الخطي مكتبة

## 1 - Import liberaries استيراد المكتبات

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

## 2 - Load the data تحميل البيانات

```python
data= np.loadtxt('data.txt',delimiter=',')
X=data[:,0].reshape(-1, 1)
y=data[:,1]
del data
```

## 3 - Plotting the data تمثيل المعطيات على شكل بياني

```python
def plotData(X,y):
    plt.figure(figsize=(8,5))
    plt.plot(X,y,'rx',markersize=5)
    plt.ylabel('Profit in $10,000s')
    plt.xlabel('Population of City in 10,000s')
    plt.show()

plotData(X,y)
```

## 4 - Create linear model  إنشاء نموذج خطي

```python
lm = linear_model.LinearRegression()
```

## 5 - Training تدريب النموذج

```python
model = lm.fit(X,y)
```

## 6 - Prediction توقع قيم جديدة

```python
value_1 = 15
value_2 = 7.5
print('For population = 150,000, we predict a profit of ','%0.2f'%(lm.predict(np.array([value_1]).reshape(-1,1))*10000))
print('For population = 75,000, we predict a profit of ','%0.2f'%(lm.predict(np.array([value_2]).reshape(-1,1))*10000))
```

### 47 - Plotting the fit line رسم الخط المستقيم

```python
plt.figure(figsize=(8,5))
plt.plot(X,y,'rx',markersize=5, label='Training Data')
plt.plot(X, lm.predict(X), color='blue', linewidth=1, label='Model: y = %0.2f + %0.2fx'%(lm.intercept_,lm.coef_))
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.legend()
plt.show()
```
